```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
% matplotlib inline
```

```python
Movie_data = pd.read_csv('Movie-Ratings.csv')
```

```python
Movie_data.info()
```

```python
# renaming columns
Movie_data.columns = ['Film','Genre','RTratings','AudienceRatings','BudgetinMM','ReleaseYear']
```

```python
#Changing datatype of columns to category
Movie_data.Film = Movie_data.Film.astype('category')
Movie_data.Genre = Movie_data.Genre.astype('category')
Movie_data.ReleaseYear = Movie_data.ReleaseYear.astype('category')
```

```python
Movie_data.info()
```

```python
Movie_data.describe().transpose()
```

```python
# creating a facet Grid
```

```python
f = sns.FacetGrid(Movie_data,row='Genre', col='ReleaseYear',hue='Genre')
kws = dict(s = 50, linewidth = 0.5, edgecolor = 'black')
f = f.map(plt.scatter, 'RTratings','AudienceRatings',**kws)
f.set(xlim = (0,100),ylim =(0,100))
for ax in f.axes.flat:
    ax.plot((0,100),(0,100), c = "gray", ls = '-')
f.add_legend()
```
